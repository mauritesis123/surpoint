# 1.1 Daily Price
## 1.1.1 Stock and stock index

Daily price is what you see on the Historial Data tab. Note that you can save the data for later use, but due to backward price adjustment, the Adj Close is likely to change in the future.


````
from datetime import datetime
import yfinance as yf
import mplfinance as mpf
start_date = datetime(2000,1,2)
end_date = datetime(2018,11,20)
start_date1 = datetime(2021, 1, 1)
end_date1 = datetime(2021, 2, 1)
data = yf.download('FALABELLA.SN', start=start_date, end=end_date)
data1 = yf.download('FALABELLA.SN', start=start_date1, end=end_date1)
mpf.plot(data1,type='candle',mav=(3,6,9),volume=True,show_nontrading=True)
````
